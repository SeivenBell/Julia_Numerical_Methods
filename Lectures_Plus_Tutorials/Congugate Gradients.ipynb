```julia
using LinearAlgebra
```

```julia
function conjugate_gradients(A, b, n)
    x0 = A \ b
    x = zero(b)
    r = b - A*x
    s = r
    for i in 1:n
        println("iter $i   |e|=$(norm(x-x0))   |r|=$(norm(r))")
        α = (r' * r) / (s' * A * s)
        x1 = x + α * s
        # r1 = r - α * A * s
        r1 = r - A * x1
        β1 = (r1' * r1) / (r' * r)
        s1 = r1 + β1 * s
        x = x1
        r = r1
        s = s1
    end
    return x
end
```

```julia
A = randn(10, 10)
A = A + A'
A = A*A
```

```julia
b = randn(10)
```

```julia
x0 = A\b
```

```julia
x = conjugate_gradients(A, b, 20)
norm(x - x0), norm(b - A*x)
```
